```python
import os
from PyPDF2 import PdfReader
import pandas as pd
```

```python
# Ruta de acceso a los archivos PDF
ruta = r"C:/Users/villa/OneDrive/Escritorio/SuperSociedades/Muestra final pre"
```

```python
def leer_pdf(ruta_archivo):
    texto = ""
    with open(ruta_archivo, "rb") as archivo:
        lector = PdfReader(archivo)
        num_paginas = len(lector.pages)
        print(f"Leyendo {num_paginas} página(s) del archivo: {ruta_archivo}")
        for pagina in lector.pages:
            texto += pagina.extract_text()
    return texto
```

```python
# Lista para almacenar los datos de cada archivo PDF
datos_pdf = []

# Iterar sobre cada archivo PDF en la ruta especificada
for archivo in os.listdir(ruta):
    if archivo.endswith(".pdf"):
        ruta_completa = os.path.join(ruta, archivo)
        # Leer el contenido del PDF como cadena de texto
        contenido_texto = leer_pdf(ruta_completa)
        # Agregar el nombre del archivo y su contenido a la lista
        datos_pdf.append((archivo, contenido_texto))
```

```python
# Crear un DataFrame de pandas con los datos recopilados
datos_completos = pd.DataFrame(datos_pdf, columns=["Nombre del archivo", "Contenido"])
```

```python
# Imprimir el DataFrame completo
print(datos_completos)
```

```python
# Agregar una nueva columna al DataFrame que contenga el número de palabras en el contenido
datos_completos['Número de palabras'] = datos_completos['Contenido'].apply(lambda x: len(x.split()))

# Imprimir el DataFrame completo
print(datos_completos)
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.util import ngrams
import networkx as nx
import matplotlib.pyplot as plt

# Descargar las stopwords en español si no están ya descargadas
nltk.download('stopwords')
nltk.download('punkt')

# Obtener las stopwords en español
stop_words = set(stopwords.words('spanish'))

# Función para realizar el preprocesamiento
def preprocesar_texto(texto):
    # Convertir a minúsculas
    texto = texto.lower()
    
    # Eliminar signos de puntuación y otros símbolos
    texto = re.sub(r'[^\w\s]', '', texto)
        
    # Eliminar números
    texto = re.sub(r'\d+', '', texto)
    
    # Eliminar palabras específicas
    palabras_eliminar = set([
       'bogota', 'arbitraje', 'centro','arbitramento', 'conciliación', 'antioquia', 'medellín', 'laudo', 'cámara', 'comercio', 'laudo', 'arbitral'
    ])
    
    # Unir stopwords y palabras específicas a eliminar
    palabras_eliminar = stop_words.union(palabras_eliminar)
    
    texto = ' '.join([palabra for palabra in texto.split() if palabra not in palabras_eliminar])
    
    # Eliminar espacios en blanco adicionales
    texto = ' '.join(texto.split())
    
    return texto
```

```python
# Aplicar el preprocesamiento a la columna "Contenido"
datos_completos['Contenido preprocesado'] = datos_completos['Contenido'].apply(preprocesar_texto)

# Imprimir el DataFrame con la columna preprocesada
print(datos_completos)
```

```python
# Función para contar el total de palabras en cada documento
def contar_total_palabras_documento(texto):
    return len(texto.split())

# Aplicar la función a la columna "Contenido preprocesado"
datos_completos['Total de palabras'] = datos_completos['Contenido preprocesado'].apply(contar_total_palabras_documento)

# Imprimir el DataFrame con el conteo total de palabras por cada documento
print(datos_completos[['Nombre del archivo', 'Total de palabras']])
```

```python
from nltk.util import ngrams

# Función para generar n-gramas de longitud 5
def generar_ngramas(tokens):
    return list(ngrams(tokens, 5))

# Aplicar la función a la columna "Tokens"
datos_completos['N-gramas'] = datos_completos['Tokens'].apply(generar_ngramas)

# Imprimir el DataFrame con las columnas requeridas
print(datos_completos[['Nombre del archivo', 'Tokens', 'N-gramas']])
```

```python
# Función para extraer la sección del texto a partir de "resuelve"
def extraer_resuelve(texto):
    # Buscar la palabra "resuelve" y extraer el texto a partir de ahí
    match = re.search(r'\bresuelve\b', texto)
    if match:
        return texto[match.start():]
    else:
        return ""

# Extraer la sección a partir de "resuelve"
datos_completos['Sección resuelve'] = datos_completos['Contenido preprocesado'].apply(extraer_resuelve)

# Imprimir las primeras filas del DataFrame
print(datos_completos[['Nombre del archivo', 'Contenido preprocesado', 'Sección resuelve']].head())
```

```python
import nltk
from nltk.util import ngrams
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_selection import SelectPercentile, chi2
import pandas as pd

# Asegúrate de tener descargado el tokenizer de nltk
nltk.download('punkt')

# Supongamos que tienes un DataFrame 'datos_completos' con una columna 'Sección resuelve'

# Función para tokenizar el texto en trigramas
def tokenizar_trigramas(texto):
    tokens = nltk.word_tokenize(texto)
    trigramas = list(ngrams(tokens, 3))
    return [' '.join(trigrama) for trigrama in trigramas]

# Aplicar la tokenización de trigramas a la columna 'Sección resuelve'
datos_completos['Trigramas'] = datos_completos['Sección resuelve'].apply(tokenizar_trigramas)

# Unir trigramas en una sola cadena de texto por documento
datos_completos['Trigramas_texto'] = datos_completos['Trigramas'].apply(lambda x: ' '.join(x))

# Definir una función para calcular los trigramas más relevantes por documento
def obtener_trigramas_relevantes(trigramas_texto, etiquetas, percentile=10):
    if not trigramas_texto.strip():  # Verifica si el texto de trigramas está vacío
        return []
    
    vectorizer = CountVectorizer()
    X = vectorizer.fit_transform([trigramas_texto])
    
    # Crear etiquetas dummy para cumplir con la función SelectPercentile
    y = [etiquetas]
    
    selector = SelectPercentile(chi2, percentile=percentile)
    X_new = selector.fit_transform(X, y)
    
    selected_features = selector.get_support(indices=True)
    selected_trigrams = [vectorizer.get_feature_names_out()[i] for i in selected_features]
    
    return selected_trigrams

# Aplicar la función de trigramas relevantes documento a documento
datos_completos['Trigramas_relevantes'] = datos_completos.apply(
    lambda row: obtener_trigramas_relevantes(row['Trigramas_texto'], 1), axis=1
)

# Mostrar el DataFrame con las nuevas columnas de trigramas
print(datos_completos[['Nombre del archivo', 'Trigramas', 'Trigramas_texto', 'Trigramas_relevantes']].head())
```
