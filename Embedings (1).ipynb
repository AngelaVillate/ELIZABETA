```python
import pandas as pd
from sentence_transformers import SentenceTransformer
```

```python
# Ruta del archivo Excel en Google Colab
ruta_archivo = "/content/sample_data/muestra101.xlsx"
df = pd.read_excel(ruta_archivo)
```

```python
# Imprimir las columnas disponibles en el DataFrame
print(df.columns)

# Desmontar el Google Drive si se ha montado
drive.flush_and_unmount()
```

```python
# Extraer la columna "Hechos" (asegúrate del nombre exacto)
hechos_column_name = 'Hechos '  # Nombre corregido si tiene un espacio al final
hechos = df[hechos_column_name].tolist()
```

```python
# Cargar el modelo de embeddings semánticos específico para español
modelo = SentenceTransformer('distiluse-base-multilingual-cased-v2')
```

```python
# Generar los embeddings
embeddings = modelo.encode(hechos, convert_to_tensor=True)
```

```python
# Convertir los embeddings a una lista de arrays de numpy
embeddings_list = embeddings.cpu().numpy().tolist()
```

```python
# Agregar los embeddings como una nueva columna al DataFrame
df['hechos-emb'] = embeddings_list
```

```python
# Guardar el DataFrame actualizado en un nuevo archivo Excel
df.to_excel("/content/sample_data/muestra101_con_embeddings.xlsx", index=False)
```

```python
# Mostrar mensaje de éxito
print("Embeddings generados y guardados correctamente en muestra101_con_embeddings.xlsx")
```

**Embeddings pretensiones**

```python
# Extraer la columna "Pretensiones" (asegúrate del nombre exacto)
pretensiones_column_name = 'Pretensiones'
pretensiones = df[pretensiones_column_name].tolist()

# Cargar el modelo de embeddings semánticos específico para español
modelo = SentenceTransformer('distiluse-base-multilingual-cased-v2')

# Generar los embeddings
embeddings = modelo.encode(pretensiones, convert_to_tensor=True)

# Convertir los embeddings a una lista de arrays de numpy
embeddings_list = embeddings.cpu().numpy().tolist()

# Agregar los embeddings como una nueva columna al DataFrame
df['pretensiones-emb'] = embeddings_list

# Guardar el DataFrame actualizado en un nuevo archivo Excel
df.to_excel("/content/sample_data/muestra101_con_embeddings_pretensiones.xlsx", index=False)

# Mostrar mensaje de éxito
print("Embeddings generados y guardados correctamente en muestra101_con_embeddings_pretensiones.xlsx")
```

```python
# Definir la columna 'Consideraciones 1'
consideraciones_columna = 'Consideraciones 1'

# Cargar el modelo de embeddings semánticos específico para español
modelo = SentenceTransformer('distiluse-base-multilingual-cased-v2')

# Extraer los textos de la columna 'Consideraciones 1'
consideraciones = df[consideraciones_columna].tolist()

# Generar los embeddings para la columna 'Consideraciones 1'
embeddings = modelo.encode(consideraciones, convert_to_tensor=True)

# Convertir los embeddings a una lista de arrays de numpy
embeddings_list = embeddings.cpu().numpy().tolist()

# Agregar los embeddings como una nueva columna al DataFrame
df[f'{consideraciones_columna}-emb'] = embeddings_list

# Guardar el DataFrame actualizado en un nuevo archivo Excel
df.to_excel("/content/sample_data/muestra101_con_embeddings_consideraciones1.xlsx", index=False)

# Mostrar mensaje de éxito
print(f"Embeddings generados y guardados correctamente en muestra101_con_embeddings_{consideraciones_columna}.xlsx")
```
